```python
# Gender-wise Sales Analysis (Warning-free + Values on Bars)
gender_sales = df.groupby('Gender')['Amount'].sum().reset_index()

plt.figure(figsize=(6,4))
ax = sns.barplot(x='Gender', y='Amount', data=gender_sales, hue="Gender", 
                 dodge=False, legend=False, palette={"F": "#FF9999", "M": "#66B2FF"})

# Add values on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge')

plt.title("Gender-wise Sales Analysis")
plt.show()
```

```python
ax = sns.countplot(x='Gender', data=df, hue='Gender',
                   dodge=False, legend=False,
                   palette={"F": "#FF9999", "M": "#66B2FF"})

for container in ax.containers:
    ax.bar_label(container)

```

```python
# Set the figure size for better visualization
plt.figure(figsize=(8,5))

# Create a countplot of Age Group distribution with Gender as hue (Male/Female)
ax = sns.countplot(x='Age Group', data=df, hue='Gender',
                   palette={"F": "#FF9999", "M": "#66B2FF"})  # Female = Pink, Male = Blue

# Add values on top of each bar
for container in ax.containers:
    ax.bar_label(container)  # Automatically labels each bar with its height (count)

# Set the chart title
plt.title("Age Group Distribution by Gender")

# Show the plot
plt.show()

```

```python
# Age Group wise total spending (Amount)
age_spending = df.groupby('Age Group')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

print(age_spending)

# Plot
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Age Group', y='Amount', data=age_spending, color="skyblue")

# Add values on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.title("Total Spending by Age Group")
plt.show()
```

```python
# State-wise Top 10 Revenue
state_revenue = df.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)

# Plot with unique colors per bar (no warning)
plt.figure(figsize=(10,6))
colors = sns.color_palette("Set2", n_colors=len(state_revenue))  # unique colors
ax = plt.bar(state_revenue['State'], state_revenue['Amount'], color=colors)

# Add values on top of bars
for i, v in enumerate(state_revenue['Amount']):
    plt.text(i, v + 100, str(int(v)), ha='center', fontsize=9)

plt.title("Top 10 States by Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Group data by Marital_Status for total spending
marital_spending = df.groupby('Marital_Status')['Amount'].sum().reset_index()
marital_spending['Marital_Status'] = marital_spending['Marital_Status'].map({0: 'Single', 1: 'Married'})

# Step 2: Plot bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(marital_spending['Marital_Status'], marital_spending['Amount'], 
               color=["#66B2FF", "#FF9999"])

# Add labels automatically on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, 
             str(int(height)), ha='center', va='bottom')

plt.title("Total Spending by Marital Status")
plt.ylabel("Amount Spent")
plt.show()
```

```python
# Step 1: Group by Marital_Status + Gender for total spending
marital_gender_spending = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index()

# Map 0/1 to labels
marital_gender_spending['Marital_Status'] = marital_gender_spending['Marital_Status'].map({0: 'Single', 1: 'Married'})

# Step 2: Plot bar chart
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Marital_Status', y='Amount', hue='Gender',
                 data=marital_gender_spending, palette={"M": "#66B2FF", "F": "#FF9999"})

# Add values on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.title("Total Spending by Marital Status and Gender")
plt.ylabel("Amount Spent")
plt.show()
```

```python
df['Marital_Label'] = df['Marital_Status'].map({0:'Single',1:'Married'})

plt.figure(figsize=(6,4))
ax = sns.countplot(x='Marital_Label', hue='Marital_Label', 
                   data=df, palette=["#66B2FF", "#FF9999"], legend=False)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Customer Count by Marital Status")
plt.show()
```

```python
# Step 1: Group by Occupation for total revenue
occupation_revenue = df.groupby('Occupation')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

# Step 2: Plot bar chart
plt.figure(figsize=(14,6))
bars = plt.bar(occupation_revenue['Occupation'], occupation_revenue['Amount'], 
               color=sns.color_palette("Set2", n_colors=len(occupation_revenue)))

# Add labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, 
             str(int(height)), ha='center', va='bottom')

plt.title("Occupation-wise Revenue")
plt.ylabel("Total Revenue (Amount)")
plt.xlabel("Occupation")
plt.xticks(rotation=45)
plt.show()
```
